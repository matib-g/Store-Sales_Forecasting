# Some functions to convert values in column between numbers and strings

```python
dictionaries = {}
def convert_to_digits(df, col):
    uniques = df[col].unique()
    #index = df.index
    unique_names = {}
    digits = []
    i = 0
    for uniq in uniques:
        unique_names[uniq] = i
        i = i + 1
    for name in tqdm(df[col]):
        digits = digits + [unique_names[name]]
    dictionaries[col] = unique_names
    return digits
```

```python
def get_key(d, value):
    return [key for key, val in d.items() if val==value]
    
def convert_back(df, col):
    unique_names = dictionaries[col]
    names = []
    for digit in tqdm(df[col]):
        names = names + get_key(unique_names, digit)
    return names
```

```python
def seasonal_plot(X, y, period, freq, ax=None):
    if ax is None:
        _, ax = plt.subplots()
    palette = sns.color_palette(
        "husl",
        n_colors=X[period].nunique(),
    )
    ax = sns.lineplot(
        x=freq,
        y=y,
        hue=period,
        data=X,
        ci=False,
        ax=ax,
        palette=palette,
        legend=False,
    )
    ax.set_title(f"Seasonal Plot ({period}/{freq})")
    for line, name in zip(ax.lines, X[period].unique()):
        y_ = line.get_ydata()[-1]
        ax.annotate(
            name,
            xy=(1, y_),
            xytext=(6, 0),
            color=line.get_color(),
            xycoords=ax.get_yaxis_transform(),
            textcoords="offset points",
            size=14,
            va="center",
        )
    return ax
```

Source: https://github.com/Kaggle/learntools/blob/master/learntools/time_series/utils.py

```python
seasonal_plot(X, y='sales', period='week', freq='day');
```

```python
def plot_periodogram(ts, detrend='linear', ax=None):
    from scipy.signal import periodogram
    fs = pd.Timedelta("1Y") / pd.Timedelta("1D")
    freqencies, spectrum = periodogram(
        ts,
        fs=fs,
        detrend=detrend,
        window="boxcar",
        scaling='spectrum',
    )
    if ax is None:
        _, ax = plt.subplots()
    ax.step(freqencies, spectrum, color="purple")
    ax.set_xscale("log")
    ax.set_xticks([1, 2, 4, 6, 12, 26, 52, 104])
    ax.set_xticklabels(
        [
            "Annual (1)",
            "Semiannual (2)",
            "Quarterly (4)",
            "Bimonthly (6)",
            "Monthly (12)",
            "Biweekly (26)",
            "Weekly (52)",
            "Semiweekly (104)",
        ],
        rotation=30,
    )
    ax.ticklabel_format(axis="y", style="sci", scilimits=(0, 0))
    ax.set_ylabel("Variance")
    ax.set_title("Periodogram")
    return ax
```

Source: https://github.com/Kaggle/learntools/blob/master/learntools/time_series/utils.py

# Investigating autocorelation

```python
def lagplot(x, y=None, shift=1, standardize=False, ax=None, **kwargs):
    from matplotlib.offsetbox import AnchoredText
    x_ = x.shift(shift)
    if standardize:
        x_ = (x_ - x_.mean()) / x_.std()
    if y is not None:
        y_ = (y - y.mean()) / y.std() if standardize else y
    else:
        y_ = x
    corr = y_.corr(x_)
    if ax is None:
        fig, ax = plt.subplots()
    scatter_kws = dict(
        alpha=0.75,
        s=3,
    )
    line_kws = dict(color='C3', )
    ax = sns.regplot(x=x_,
                     y=y_,
                     scatter_kws=scatter_kws,
                     line_kws=line_kws,
                     lowess=True,
                     ax=ax,
                     **kwargs)
    at = AnchoredText(
        f"{corr:.2f}",
        prop=dict(size="large"),
        frameon=True,
        loc="upper left",
    )
    at.patch.set_boxstyle("square, pad=0.0")
    ax.add_artist(at)
    title = f"Lag {shift}" if shift > 0 else f"Lead {shift}"
    ax.set(title=f"Lag {shift}", xlabel=x_.name, ylabel=y_.name)
    return ax
```

Source: https://github.com/Kaggle/learntools/blob/master/learntools/time_series/utils.py

```python
def plot_lags(x,
              y=None,
              lags=6,
              leads=None,
              nrows=1,
              lagplot_kwargs={},
              **kwargs):
    import math
    kwargs.setdefault('nrows', nrows)
    orig = leads is not None
    leads = leads or 0
    kwargs.setdefault('ncols', math.ceil((lags + orig + leads) / nrows))
    kwargs.setdefault('figsize', (kwargs['ncols'] * 2, nrows * 2 + 0.5))
    fig, axs = plt.subplots(sharex=True, sharey=True, squeeze=False, **kwargs)
    for ax, k in zip(fig.get_axes(), range(kwargs['nrows'] * kwargs['ncols'])):
        k -= leads + orig
        if k + 1 <= lags:
            ax = lagplot(x, y, shift=k + 1, ax=ax, **lagplot_kwargs)
            title = f"Lag {k + 1}" if k + 1 >= 0 else f"Lead {-k - 1}"
            ax.set_title(title, fontdict=dict(fontsize=14))
            ax.set(xlabel="", ylabel="")
        else:
            ax.axis('off')
    plt.setp(axs[-1, :], xlabel=x.name)
    plt.setp(axs[:, 0], ylabel=y.name if y is not None else x.name)
    fig.tight_layout(w_pad=0.1, h_pad=0.1)
    return fig
```

Source: https://github.com/Kaggle/learntools/blob/master/learntools/time_series/utils.py

```python
plot_pacf(y_deseason, lags=8);
plot_lags(y_deseason, lags=8, nrows=2);
```

There is no autocorelation in summary sales, but let's check if there is autocorelation in specyfic group of products

```python
supply_sales = family_sales.loc(axis=1)[:, 'SCHOOL AND OFFICE SUPPLIES']
y_family = supply_sales.loc[:, 'sales'].squeeze()

fourier = CalendarFourier(freq='M', order=4)
dp = DeterministicProcess(
    constant=True,
    index=y_family.index,
    order=1,
    seasonal=True,
    drop=True,
    additional_terms=[fourier],
)
X_time = dp.in_sample()
X_time['NewYearsDay'] = (X_time.index.dayofyear == 1)

model = LinearRegression(fit_intercept=False)
model.fit(X_time, y_family)
y_family_deseason = y_family - model.predict(X_time)
y_family_deseason.name = 'family_sales_deseasoned'
```

```python
plot_pacf(y_family_deseason, lags=8);
plot_lags(y_family_deseason, lags=8, nrows=2);
```

We can see that there is autocorelation in sales in SCHOOL AND OFFICE SUPPLIES family and probably not only in this family. The correlogram indicates the first lag is likely to be significant, as well as possibly the eighth lag. The lag plot suggests the effect is mostly linear.

Lets's check if 'onpromotion' is good leading indicator.

```python
onpromotion = supply_sales.loc[:, 'onpromotion'].squeeze().rename('onpromotion')

# Drop days without promotions
plot_lags(x=onpromotion.loc[onpromotion > 1], y=y_family_deseason.loc[onpromotion > 1], lags=3, leads=3, nrows=1);
```

The lag plot indicates that both leading and lagged values of onpromotion are correlated with supply sales. This suggests that both kinds of values could be useful as features. There may be some non-linear effects as well.

### Forcasting with lags

```python
def make_lags(ts, lags, lead_time=1, name='y'):
    return pd.concat(
        {
            f'{name}_lag_{i}': ts.shift(i)
            for i in range(lead_time, lags + lead_time)
        },
        axis=1)
```

Source: https://github.com/Kaggle/learntools/blob/master/learntools/time_series/utils.py

```python
def make_leads(ts, leads, name='y'):
    return pd.concat(
        {f'{name}_lead_{i}': ts.shift(-i)
         for i in reversed(range(leads))},
        axis=1)
```

Source: https://github.com/Kaggle/learntools/blob/master/learntools/time_series/utils.py

```python
X_lags = make_lags(y_family_deseason, lags=1)

X_promo = pd.concat([
    make_lags(onpromotion, lags=1, name='y_promo'),
    onpromotion,
    make_leads(onpromotion, leads=1),
], axis=1)

X_family = pd.concat([X_time, X_lags, X_promo], axis=1).dropna()
y_family, X_family = y_family.align(X_family, join='inner')
```

```python
X_family_train, X_family_valid, y_family_train, y_family_valid = train_test_split(X_family, y_family, test_size=30, shuffle=False)

model = LinearRegression(fit_intercept=False).fit(X_family_train, y_family_train)
y_family_fit = pd.Series(model.predict(X_family_train), index=X_family_train.index).clip(0.0)
y_family_pred = pd.Series(model.predict(X_family_valid), index=X_family_valid.index).clip(0.0)

rmsle_train = mean_squared_log_error(y_family_train, y_family_fit) ** 0.5
rmsle_valid = mean_squared_log_error(y_family_valid, y_family_pred) ** 0.5
print(f'Training RMSLE: {rmsle_train:.5f}')
print(f'Validation RMSLE: {rmsle_valid:.5f}')

ax = y_family.plot(**plot_params, alpha=0.5, title="Average Sales in SCHOOL AND OFFICE SUPPLIES", ylabel="items sold")
ax = y_family_fit.plot(ax=ax, label="Fitted", color='C0')
ax = y_family_pred.plot(ax=ax, label="Forecast", color='C3')
ax.legend();
```

### Forecasting with additional rolling window parameters

```python
y_family_lag = supply_sales.loc[:, 'sales'].shift(1)
y_family_lag = y_family_lag.iloc[1:]
onpromo = supply_sales.loc[:, 'onpromotion']
onpromo = onpromo.iloc[1:]

# 28-day mean of lagged target
rolling_mean_7 = y_family_lag.rolling(window=7, min_periods=1).mean()
# 14-day median of lagged target
rolling_median_14 = y_family_lag.rolling(window=14, min_periods=1).median()
# 7-day rolling standard deviation of lagged target
rolling_std_7 = y_family_lag.rolling(window=7, min_periods=1).std()
# 7-day sum of promotions with centered window
rolling_promo_7 = onpromo.rolling(window=7, center=True, min_periods=1).sum()
```

```python
rolling_col_names = ['rolling_mean_7','rolling_median_14','rolling_std_7','rolling_promo_7']
rolling_metrics = pd.concat([rolling_mean_7,rolling_median_14,rolling_std_7,rolling_promo_7], axis=1)
rolling_metrics.columns = rolling_col_names
rolling_metrics = rolling_metrics.fillna(method='backfill')
rolling_X_family = pd.concat([X_family,rolling_metrics], axis=1)
```

```python
X_family_train, X_family_valid, y_family_train, y_family_valid = train_test_split(rolling_X_family, y_family, test_size=30, shuffle=False)

model = LinearRegression(fit_intercept=False).fit(X_family_train, y_family_train)
y_family_fit = pd.Series(model.predict(X_family_train), index=X_family_train.index).clip(0.0)
y_family_pred = pd.Series(model.predict(X_family_valid), index=X_family_valid.index).clip(0.0)
rolling_fit = y_family_fit
rolling_pred = y_family_pred

rmsle_train = mean_squared_log_error(y_family_train, y_family_fit) ** 0.5
rmsle_valid = mean_squared_log_error(y_family_valid, y_family_pred) ** 0.5
print(f'Training RMSLE: {rmsle_train:.5f}')
print(f'Validation RMSLE: {rmsle_valid:.5f}')

ax = y_family.plot(**plot_params, alpha=0.5, title="Average Sales in SCHOOL AND OFFICE SUPPLIES", ylabel="items sold")
ax = y_family_fit.plot(ax=ax, label="Fitted", color='C0')
ax = y_family_pred.plot(ax=ax, label="Forecast", color='C3')
ax.legend();
```

```python
rmsle_train = mean_squared_log_error(y_family_train, finall_fit) ** 0.5
rmsle_valid = mean_squared_log_error(y_family_valid, finall_pred) ** 0.5
print(f'Training RMSLE: {rmsle_train:.5f}')
print(f'Validation RMSLE: {rmsle_valid:.5f}')

ax = y_family.plot(**plot_params, alpha=0.5, title="Average Sales in SCHOOL AND OFFICE SUPPLIES", ylabel="items sold")
ax = finall_fit.plot(ax=ax, label="Fitted", color='C0')
ax = finall_pred.plot(ax=ax, label="Forecast", color='C3')
```
